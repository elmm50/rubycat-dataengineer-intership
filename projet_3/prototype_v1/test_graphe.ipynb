```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

  

x_axis=["Lundi","Mardi","Mercredi","Jeudi","Vendredi","Samedi","Dimanche"]
y_axis= [8,7,8,11,6,3,6]

y1_limit=[10,8,9,10,8,3,3]
y2_limit=[6,4,6,5,6,1,0]

# plot
plt.figure(figsize=(14, 8))
plt.title("Nombre d'heures de connexion de l'Utilisateur MB45")
plt.ylabel("Nombre d'heures de connexion")
plt.xlabel("Jour")
plt.plot(x_axis,y1_limit,color='blue')
plt.plot(x_axis, y_axis, linestyle='-', marker='o',color='red',label='Usage courant')
plt.plot(x_axis,y2_limit,color='blue')
plt.fill_between(x_axis, y1_limit, y2_limit, color='blue', alpha=0.3,label='Usage normal')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Donnes pour les courbes
x = ['01/01/2022','02/01/2022','03/01/2022','04/01/2022','05/01/2022','06/01/2022','07/01/2022','08/01/2022']
y1 = np.array([0.5,0.1,0.4,0.6,0.8,0.4,0.3,0.5])


plt.figure(figsize=(14, 8))
plt.title("Reconstruction error de l'Utilisateur MB45")
plt.ylabel("Reconstruction error")
plt.xlabel("Jour")
# Tracer les courbes
plt.bar(x, y1,label='Reconstruction error')
plt.axhline(y=0.7, color='r', linestyle='--',label='Threshold')



# Afficher la lgende et le graphe
plt.legend()
plt.show()

```
